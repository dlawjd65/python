**Padndas 기초**
pandas는 표형태의 데이터를 다루는 라이브러리

```
import pandas as pd

s = pd.Series([10, 20, 30, 40, 50])
print(s)
```

```
h = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
print(h)
print(h['b'])
```

```
data = {
    'name' : ['Kim', 'Lee', 'Park', 'Choi'],
    'age' : [25, 30, 35, 28],
    'city' : ['Seoul', 'Busan', 'Seoul', 'Daegu']
}
df = pd.DataFrame(data)
print(df)
```

```
students = pd.DataFrame({
    'name':['김철수', '이영희', '박민수'],
    'age': [20, 21, 22],
    'grade':['A', 'B', 'A']
})

print(students)
print("/n컬럼:", students.colums.tolist())
```
